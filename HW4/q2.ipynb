```python
from cmdstanpy import CmdStanModel
```

```python
# Simulated data based on the class recognition task
stan_data = {
    "N_old": 10,
    "N_new": 10,
    "y_old_old": 8,   # hits
    "y_new_old": 3    # false alarms
}
```

```python
model_1ht = CmdStanModel(stan_file="q2_1HT.stan")
fit_1ht = model_1ht.sample(
    data=stan_data,
    chains=4,
    parallel_chains=4,
    iter_warmup=1000,
    iter_sampling=1000,
    seed=42
)
```

```python
summary_1ht = fit_1ht.summary()
print("1HT summary:")
print(summary_1ht)
```

```python
model_2ht = CmdStanModel(stan_file="q2_2HT.stan")
fit_2ht = model_2ht.sample(
    data=stan_data,
    chains=4,
    parallel_chains=4,
    iter_warmup=1000,
    iter_sampling=1000,
    seed=42
)
```

```python
summary_2ht = fit_2ht.summary()
print("2HT summary:")
print(summary_2ht)
```
