```python
import pandas as pd
import numpy as np
from cmdstanpy import CmdStanModel
```

```python
df = pd.read_csv("response_times.csv",sep=";")
print(df.head())
print(df.columns)
```

```python
unique_ids = sorted(df["id"].unique())
id_map = {old: new for new, old in enumerate(unique_ids, start=1)}
df["id_stan"] = df["id"].map(id_map)

stan_data = {
    "N": len(df),
    "J": len(unique_ids),
    "id": df["id_stan"].astype(int).tolist(),
    "y": df["rt"].astype(float).tolist(),
    "condition": df["condition"].astype(int).tolist(),
    "choice": df["choice"].astype(int).tolist(),
}
```

```python
model = CmdStanModel(stan_file="diffusion_multiple.stan")
```

```python
fit = model.sample(
    data=stan_data,
    chains=4,
    parallel_chains=4,
    iter_warmup=1000,
    iter_sampling=1000,
    seed=42
)

summary = fit.summary()
print(summary)

```

```python

draws = fit.draws_pd()

v1_cols = [c for c in draws.columns if c.startswith("v_cond1[")]
v2_cols = [c for c in draws.columns if c.startswith("v_cond2[")]

v1_person_means = draws[v1_cols].mean(axis=0)
v2_person_means = draws[v2_cols].mean(axis=0)

print("\nCondition 1 drift means by person:")
print(v1_person_means)

print("\nCondition 2 drift means by person:")
print(v2_person_means)

draw_mean_v1 = draws[v1_cols].mean(axis=1)
draw_mean_v2 = draws[v2_cols].mean(axis=1)

print("\nPosterior mean drift, condition 1:", draw_mean_v1.mean())
print("Posterior mean drift, condition 2:", draw_mean_v2.mean())

print("P(cond1 drift < cond2 drift | data) =", np.mean(draw_mean_v1 < draw_mean_v2))
print("P(cond2 drift < cond1 drift | data) =", np.mean(draw_mean_v2 < draw_mean_v1))

drift_rows = [idx for idx in summary.index if "v_cond" in idx]
print()
print()
print(summary.loc[drift_rows, ["R_hat", "ESS_bulk"]])
```
