```python
import numpy as np
import pandas as pd
from cmdstanpy import CmdStanModel
```

```python
np.random.seed(42)

N = 100
alpha_true = 2.0
beta_true = 3.5
sigma_true = 1.0

x = np.random.normal(size=N)
y = alpha_true + beta_true * x + sigma_true * np.random.normal(size=N)

data = {
    "N": N,
    "x": x,
    "y": y
}
```

```python
model = CmdStanModel(stan_file="q5.stan")
fit = model.sample(
    data=data,
    chains=4,
    iter_sampling=1000,
    iter_warmup=1000,
    seed=42
)
```

```python
print(fit.summary())
```

```python
N_big = 1000
x_big = np.random.normal(size=N_big)
y_big = alpha_true + beta_true * x_big + sigma_true * np.random.normal(size=N_big)

data_big = {
    "N": N_big,
    "x": x_big,
    "y": y_big
}

fit_big = model.sample(
    data=data_big,
    chains=4,
    iter_sampling=1000,
    iter_warmup=1000,
    seed=42
)

```

```python
print(fit_big.summary())
```
